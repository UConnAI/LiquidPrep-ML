```python
#importing libraries
import requests
import json
```

```python
#url is for precipitation data
url = "https://api.synopticlabs.org/v2/stations/precipitation?stid=TWB{}&units=english&pmode=intervals&interval=day&start=20{}01011200&end=20{}12301200&token=f626b8d4c80242d39ecceaa253c41aff"
#url1 is for timeseries data
url1 = "https://api.synopticlabs.org/v2/stations/timeseries?stid=TWB{}&vars=precip_accum_fifteen_minute,precip_accum_five_minute,pressure,relative_humidity,soil_temp,soil_moisture,solar_radiation,air_temp,wind_direction,wind_gust,wind_speed&units=english&start=20{}01011200&end=20{}12311200&token=f626b8d4c80242d39ecceaa253c41aff"
```

```python
#l is the list of year batches that we are to extract data of
#first for loop is for the list of station IDs of Texas
#second for loop is iterating over all batches of years
#the following code extracts all json responses for respective years and station IDs

l = [["16", "17"], ["18", "19"], ["20", "21"], ["22", "22"]]
for i in range(1, 96):
    id_val = str(i)
    id_val = '0' + id_val if len(id_val) == 1 else id_val
    for year in l:
        x = requests.get(url1.format(id_val, year[0], year[1]))
        data = x.json()
        print(id_val, year, x.status_code)
        with open('data_scraped/data-{}-{}-{}.json'.format(id_val, year[0], year[1]), 'w') as f:
            json.dump(data, f)
```
